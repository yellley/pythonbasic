```python
#11.
##cut : 실수 값으 경계선을 지정하는 경우
###x=1차원 형태의 배열 형태
###bins = int, 스칼라를 요소로 갖는 시퀀스가 옴

##qcut : 개수가 똑같은 구간으로 나누는 경우(분위수)
###x=1d ndarray 혹은 series
###q = int혹은 분위수를 나타내는 1이하의 실수를 요소로 갖는 list



#예시
ages = [0, 2, 10, 21, 23, 37, 31, 61, 20, 41, 32, 101]
#컷 명령을 사용하면 실수값을 다음처럼 카테고리 값을 나눌 수 있음
#빈스 인수는 카테고리를 나누는 기준값이됨
#영역을 넘는 값은 논
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "장년", "중년", "노년"]
cats = pd.cut(ages, bins, labels = labels)
cats

#컷()명령이 반환하는 값은 categorical 클래스 객체임
#categories속성으로 label문자열을, codes속성으로 정수로 인코딩한 카테고리 값을 가짐
type(cats)
cats.categories
cats.codes



df4 = pd.DataFrame(ages, columns=["ages"])
df4["age_cat"] = pd.cut(df4.ages, bins, labels = labels)
df4.dtypes
#위 칼럼값은 문자열 아님
df4["age_cat"].astype(str) + df4["ages"].astype(str)




##qcut()명령은 구간 경계선을 지정하지 않고 분위수와 같이 데이터 개수가 같도록 구간을 나눔
##예시 : 1000개의 데이터를 4개의 구간으로 나누는데 각 구간은 250개씩
data = np.random.randn(1000)
cats = pd.qcut(data, 4, labels = ["Q1","Q2","Q3","Q4"])
cats
pd.value_counts(cats)
```
