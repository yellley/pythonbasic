```python
#연습-----------------------------------------------------고칠거 있음ㅌㅇㅌㅇㅊ
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset("titanic")
titanic.dropna(subset = ['age'], inplace = True)#nan값 배제할 때



bins = [0, 20, 30, 50, 70, 100] #pd.cut()사용, 나이범위지정해주는 역할
labels = ["미성년자", "청년", "장년", "중년", "노년"]
explode = (0.1, 0, 0, 0, 0)
colors = ["yellowgreen", "gold", "lightskyblue", "lightcoral", "orange"]


titanic_cut = pd.cut(titanic["age"], bins = bins, labels = labels).value_counts(True)#value_counts(트루)하면 비율로 나옴, 키워드인수로 들어가면 순서가 상관없는데 그게 아니라면 순서지켜야함->bins, labels
#이부분 잘 모르겠음

plt.title("타이타닉")
plt.pie(titanic_cut, explode=explode, labels = labels, colors = colors,
       autopct = '%1.1f%%', shadow = True, startangle = 90) #1:최소 한자리, .1 : 소수점 한자리 
plt.axis('equal') #원 형태 유지
plt.show()
```

```python
#imshow
##자료의 시각화를 돕기 위해 다양한 2차원 인터폴레이션을 지원

methods = [
    None, 'none', 'nearest', 'bilinear', 'bicubic', 'spline16',
    'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric',
    'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos'
]
fig, axes = plt.subplots(3, 6, figsize=(12, 6),
                         subplot_kw={'xticks': [], 'yticks': []})
for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(X, cmap=plt.cm.bone_r, interpolation=interp_method)
    ax.set_title(interp_method)
plt.show()
```

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)

plt.title("Contour plots")
plt.contourf(XX, YY, ZZ, alpha = 75, cmap = 'jet')
plt.contourf(XX, YY, ZZ, colors = 'black')
plt.show()
```

```python
#3D 서피스 플롯
from mpl_toolkits.mplot3d import Axes3D

X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y)
RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title('3D surface Plot')
ax.plot_surface(XX, YY,ZZ, rstride = 1, cstride = 1, cmap = 'hot')
plt.show()
```

```python
#카운트플롯
##class는 선실등급을 나타내는 칼럼
##마치 value_counts()의 결과를 보듯 바차트 볼 수 있음
##각 유니크한 값들이 몇개씩 있는지 시각적으로 확인 가능

df = sns.load_dataset("titanic")
sns.countplot(x = df["class"])# x에 시리즈를 넣어주고 있음, #x키워드 인수에 시리즈를 전달, 
                                #시리즈에 있는 값을 유니크하게 가져와 카운트한 결과를 가져올 수 있음
    
```

```python
#hue키워드인수
##hue활용해 분류할 두 번째 변수 설정 가능
###여기서는 alive를 칼럼으로 지정해 선실별 생존 여부를 가시적으로 확인할 수 있게 도움
sns.countplot(data = df, x ='class', hue="alive") #hue에 칼럼의 이름을 넣어줌, 값이 여러개면 여러값이 가능한대, hue랑은 한개의 변수만 추가 가능함
```

```python
##kde 커널이라는 함수를 겹치는 방법
##히스토그램보다 부드러운 형태의 분포곡선
#아래예제는 kde키워드 인수에 트루값을 전달해 히스토그램과 동시에 출력
sns.histplot(data = penguins, x = 'flipper_length_mm', kde = True)
```

```python
##hue키워드 인수로 데이터 분리하기
##hue키워드 인수에 분류 기준이 될 칼럼 레이블을 전달해서 여러 개의 히스토그램을 합친 것과 같은 결과의 그래프 출력

sns.histplot(data = penguins, x = 'flipper_length_mm', hue = "species")
```

```python
##다양한 표현법 stack
##전 예제는 layer가 겹쳐서 표현 -> 이번 예제에서는 누적하듯이
##multiple 키워드 인수에 stack값을 전달하면됨
sns.histplot(data = penguins, x = 'flipper_length_mm', hue = "species", multiple = "stack")
```

```python
##step
##오버래핑된 결과가 보기 힘들 때 해결 방법 중 하나
##element키워드 인수에 step값 전달
sns.histplot(data = penguins, x = 'flipper_length_mm', hue = "species", element = "step")
```

```python
##poly
##다각형 형태로 표현
##element키워드 인수에 ploy값 전달
sns.histplot(data = penguins, x = 'flipper_length_mm', hue = "species", element = "poly")
```

```python
##poly
##x축 y축 모두 칼럼 할당 시 히트맵형태 히스토그램
sns.histplot(data = penguins, x = 'bill_depth_mm', y = 'body_mass_g', hue = "species")
```

```python
##poly
##x축 y축 모두 칼럼을 할당할 때 , 그 중 하나의 값이 이산된 값이라면 아래와 같이 데이터를 시각화하면 훨씬 이해가 더 쉬움
sns.histplot(
    penguins, x = 'bill_depth_mm', y = 'species', hue = "species", legend = False
)
```

```python
#displot()
##1,2차원 데이터의 분포를 히스토그램으로 표현
##rug, kde에 대해 동시에 표현이 가능해서 표현 범위가 더 넓음

##histplot()과 동일한 인수와 그래프 모양
```

```python
#kde
#kind 키워드 인수 활용해 kde로 변경
sns.displot(data = penguins, x = 'flipper_length_mm', kind = 'kde')
```

```python
#ecdf
##kind키워드 인수에 ecdf -> 누적 분포 그래프
sns.displot(data = penguins, x = 'flipper_length_mm', kind = 'ecdf')
```

```python
#히스토그램과 kde 동시 표현
##kde = True
sns.displot(data = penguins, x = 'flipper_length_mm', kde = True)
```

```python
#x,y모두 값 할당하기
##x,y에 각각 값 할당해 히트맵 형태의 그래프 출력
sns.displot(data = penguins, x = 'flipper_length_mm', y='bill_length_mm', hue='species')
```

```python
#2개 값을 kde로 표현
##위 히스토그램을 kde로도 표현 가능
##rug = True키워드 인수를 전달해 rug에 대한 ㅠ표현도 동시에 가능

ㅎ = sns.displot(data = penguins, x = 'flipper_length_mm', y='bill_length_mm', kind ="kde", rug = True)
```
