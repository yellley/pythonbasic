```python
#10. DF 고급 인덱싱 -> loc[], iloc[] 

## 특정 데이터만 골라내는 것(인덱싱)
##순서 : 로우가 먼저 칼럼이 그 다음
## pandas는 콤마를 사용한 형식의 2차원 인덱싱 지원, 특별한 인덱서 속성 제공

### loc : label값 기반의 2차원 인덱싱 = 문자로 된 것들,
### iloc : 순서를 나타내는 정수 기반의 2차원 인덱싱 = 숫자로 된 것들

df.loc[row 인덱싱 값] #컬럼 생략 -> 전체 칼럼에서 볼거야!라는 뜻이라고 봐도 됨 / 인덱싱,슬라이싱 가능
df.loc[row 인덱싱 값, column 인덱싱 값]  #인덱싱/슬라이싱 가능

```

```python
#10-1. loc인덱서
## row 인덱싱 값 : 정수 or row 인덱스 데이터
## column 인덱싱 값 : 정수 or label문자열
## 종류 : 인덱스데이터, 인덱스 데이터 슬라이스, 인덱스 데이터 리스트, 같은 로우 인덱스를 가지는 boolean 시리즈(로우 인덱싱의 경우), 위의 값들을 반환하는 함수 
            ##########################다시 질문하기

###Df만들기
data = np.arange(10, 22).reshape(3,4)
# data
columns = ["A", "B", "C", "D"]
index = ["a", "b", "c"]

df = pd.DataFrame(data, columns = columns, index = index)
df #이방법말고 한꺼번에 넣고 한번에 만들수도 있음


##loc인덱서를 사용하면서 인덱스를 콤마없이 하나만 넣으면 row선택함
df.loc["a"] #인덱스 데이터가 a인 행을 고르면 해당하는 로우가 시리즈로 반환, 시리즈라서 상하로 표현되지만 로우를 가져온거임


##loc로 슬라이싱도 가능
###밑에 코드 둘이 같은 값 출력됨
df.loc['b' : 'c'] #loc사용한거
df['b' : 'c'] #그냥 슬라이싱한거


##리스트 형태로도 인덱싱 가능
###int는 iloc, 문자는 loc
df.loc[['b', 'c']] #값이 두개라 df로 표현, 슬라이싱과 다르게 loc안쓰면 키에러뜸


##boolean 시리즈로 row기준 인덱싱 가능
df.A > 15 #df.문자칼럼 <- 이것만 가능, 조건에 따라 true/false로 출력됨
df.loc[df.A > 15] #A칼럼에 15보다 큰게 c로우에 18밖에 없어서 c로우만 출력됨



##callable한 함수 만들어 인덱싱 가능
def select_rows(df, num):# A열의 값이 10보다 큰 row만 선택(출력)
    return df.A > num
    
select_rows(df, 10)

df.loc[select_rows]##################################################여기다시



## loc인덱서는 칼럼 레이블 인덳이이나 칼럼 리이블리스트 인덱싱은 불가능
df.loc['A'] #키에러 뜰거임 -> row값이아니라 column값으로 줘서 오류


##default로 주어지는 정수로 생성된 DF의 경우 슬라이싱숫자가 포함된 결과 가져옴 -> loc는 label index에 접근하기 때문
df2 = pd.DataFrame(np.arange(10, 26).reshape(4, 4), columns = ["A", "B", "C", "D"])
df2

##원래 로우 인덱스 값이 정수인 경우 -> 마지막 값이 포함됨
df2.loc[1 : 2]


###슬라이싱 할 때 마지막 것(숫자) 포함여부###
#### loc -> lable -> 포함
df2.loc[1 : 2] #row 1,2출력

### iloc -> int index -> 불포함
df2.iloc[1 : 2] #row 1출력


## 인덱싱 값을 로우/칼럼 모두 받으려면 "df.loc[row인덱스, column인덱스]"
df.loc["a", "A"] #row 인덱스 label값이 a, column인덱스 label값이 A인 위치의 값을 구해줌


## 콤마로 구분된 인덱싱 값으로 label데이터의 슬라이싱/리스트도 사용 가능
df.loc["b" : "A"]mf
df.loc["a", : ]
df.loc[["a" , "b"], ["B" , "D"]] #row a,b 와 column B, D인 값만 찾아줌


## row인덱스가 같은 boolean 시리즈나 이런 불린 시리즈를 반환하는 함수도 row의 인덱싱 값이 될 수 있음
df.loc[df.A > 10, ['C', 'D']] #칼럼 A에서 10보다 큰 로우가 있는 내용 출력
```
