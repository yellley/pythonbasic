```python
def give_me_the_awesome_map(column, bins):
    # 앞서 전처리한 데이터를 불러옵니다.
    csvFile = 'Tax/location_for_map.csv'
    SiDodf = pd.read_csv(csvFile, encoding = 'utf-8')

    # 시도 행정구역에 대한 json파일을 지정합니다.
    SiDo_geo = 'https://raw.githubusercontent.com/suanlab/dataset/master/skorea-provinces-2018-geo.json'
    
    # 기본 맵을 설정합니다.
    m = folium.Map(location = [36.45, 127.42],
                    tiles = "OpenStreetMap", zoom_start = 8)
    
    # 시도에 맞는 데이터를 맵핑하여 색상으로 인코딩한 레이어를 생성합니다.
    folium.Choropleth(
        geo_data = SiDo_geo,
        name = '업종별 분포 지도',
        data = SiDodf,
        columns = ['area', column],
        key_on='feature.properties.name_eng',
        fill_color = 'PuRd',
        fill_opacity = 0.7,
        bins=bins,
        line_opacity = 0.2,
        legent_name = "개업수").add_to(m)
    
    # 지도에 레이어 컨트롤을 추가합니다.
    folium.LayerControl().add_to(m)
    
    m.save(f'maps/{column.replace(" ", "_")}_map.html')
    # 저장한 파일을 열어봅니다.
    # webbrowser.open(f'maps/{column.replace(" ", "_")}_map.html')
```

```python
# 파일을 다시 불러옵니다.
SiDodf2 = pd.read_csv('Tax/location_for_map.csv', encoding = 'utf-8')
display(SiDodf)
```

```python
# 수치 값이 아닌 column area를 지웁니다.
del SiDodf2["area"]
```

```python
# 수치 값을 갖는 전체 컬럼에 대해 반복
for column in SiDodf.columns:
    give_me_the_awesome_map(column, 
                            # 백분위 수 중 10분위, 20분위, 30분위, 40분위, 50분위, ... 100분위까지 범위를 갖는 bins 구합니다.
                            [np.percentile(SiDodf[column], _) for _ in range(0, 101, 10)])    
```
