```python
#DF 그룹 연산 메서드

#합계 어떻게?
np.random.seed(0)
df2 = pd.DataFrame({
    'key1': ['A', 'A', 'B', 'B', 'A'],
    'key2': ['one', 'two', 'one', 'two', 'one'],
    'data1': [1, 2, 3, 4, 5],
    'data2': [10, 20, 30, 40, 50]
})
df2
#group by : 그룹A와 그룹B로 구분한 그룹 데이터 만들 수 있게 해줌
groups = df2.groupby(df2.key1)

groups.sum()



##groupby클래스 객체를 명시적으로 얻을 필요가 없을 때
df2.data1.groupby(df2.key1).sum() 


##그룹분석한 결과에서 data1만 뽑아도 됨
###밑에 두가지는 똑같은 결과값 다른방법
df2.groupby(df2.key1)["data1"].sum()
df2.groupby(df2.key1).sum()["data1"]


##복합키 key1, key2값에 따른 data1의 합계 구하기
df2.data1.groupby([df2.key1, df2.key2]).sum()
###이 결과를 unstack명령으로 피봇 테이블 형태로 만들 수 있음(만들 수 있지만 지금은 내용에서 빼심)

```

```python
#틱설정하기
##틱 : 플롯/차트에서 축상의 위치 표시 지점
## 수동으로 설정하고 싶을 때 : xticks(), yticks()
import numpy as np

X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x축과 y축의 tick label설정")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, 1])
plt.show()
```

```python
##틱 라벨 문자열에는 $$사이에 수학 문자식 삽입 가능
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("TeX, 문자열로 tick label정의")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.show()
```

```python
#여러개 선 그리기
##x데이터,y데이터,스타일 문자열을 반복해 인수로 넘김

t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--', t, 0.5* t **2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.show()



##하나의 plot()이 아닌 복수의 plot을 하나의 그림에 겹쳐서 그릴 수 있음
plt.title("복수의 plot명령을 한 그림에서 표현")
style1 = {"c":"b", "lw":5, "ls":"--", "marker":"o", "ms":15, "mec":"g", "mew":5, "mfc":"r"}
style2 = {"c":"k", "lw":3, "ls":":", "marker":"s", "ms":10, "mec":"m", "mew":5, "mfc":"c"}

plt.plot([1, 4, 9, 16], **style1) #**는 키워드 인자를
plt.plot([9, 16, 4, 1], **style2)
plt.show()
```

```python
#범례(Legend)
##여러개의 라인 플롯을 동시에 그리는 경우 각 선이 무슨 자료를 표시하는지 보여줌
##수동으로 설정 원하면 loc키워드 인수 사용


##plot마다 label키워드 인수를 통해 legend label을 설정
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)

plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls='--', label = 'cosine')
plt.plot(X, S, ls=':', label = 'sine')
plt.legend(loc=2) #upper left, 라벨 박스 위치 정해주는 곳
plt.show()
```

```python
# x, y축 레이블

X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C, label = "cosine")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Cosine Plot")
plt.show()
```

```python
#Figure
##pyplot.subplots()사용
##Axes.plot()를 활용해 Axes에 데이터를 그림

fig,ax = plt.subplots() #Create a figure containing a single axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]) # Plot some data on the axes.




##figure는 전체 그림(그래프)을 의미하는 객체
##figure 객체는 Axes, 특수 Artist그룹(title, figure legends, colorbar, etc), 중첩된 subfigures까지 포함
##Axes없이 figure만 얻기 : figure()메서드 사용
fig = plt.figure() #an empty with no Axes
fig, ax = plt.subplots() #a figure with a single Aexs -> 박스1개
fig, axs = plt.subplots(2, 2) #a figure with a 2x2 grid of Axes -> 박스 4개만들어줌
```
