```python
mpl.rcParams["figure.figsize"] = (16, 72)

# shape이 (25, 4)인 axes
fig, axs = plt.subplots(25, 4) # Figure객체, Axes를 (25, 4)의 shape으로 갖는 ndarray 객체 
# axs.shape
# print(axs)
# print(type(axs))
# axs.shape의 결과가 (25, 4)
rows, cols = axs.shape

# season DataFrame에 대한 row index에 접근하기 위한 변수
idx = 0

# 0, 1, 2, 3, ..., 24
for row in range(rows):
    # 0, 1, 2, 3
    for col in range(cols):
        sns.barplot(x=season.iloc[idx], 
                    y=season.iloc[idx].index, 
                    # axs[0, 0], axs[0, 1], axs[0, 2], axs[0, 3], axs[1, 0], axs[1, 1], ..., axs[24, 3]
                    ax=axs[row, col])
        # row index 값을 1씩 증가
        idx += 1

# grid 간에 간격을 자동 조절해줍니다.
plt.tight_layout()
plt.savefig('season.png') # 차트저장은 show위로
plt.show()
  
```
